```python
# Description of data
# Each user in the dataset is represented by a unique id (uuid).
# A person can view a document (document_id), which is simply a web page with content (e.g.  a news article). 
# On each document, a set of ads (ad_id) are displayed. 
# Each ad belongs to a campaign (campaign_id) run by an advertiser (advertiser_id).
# You are also provided metadata about the document, such as which entities are mentioned, 
# a taxonomy of categories, the topics mentioned, and the publisher.
```

```python
# the document_id in events is the document_id of the page presenting the ad_id(s) to the user.
# document_id in promoted content is document_id for landing page for a given ad_id. So if you will click to that ad_id you will see document/website with corresponding document_id.
```

```python
import pandas as pd
import os
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## train and rest data
train = pd.read_csv('../data/raw/clicks_train.csv.zip', compression='zip')
test = pd.read_csv('../data/raw/clicks_test.csv.zip', compression='zip')
```

```python
# events provides info about display id
events = pd.read_csv('../data/raw/events.csv.zip', compression='zip')
events.head()
```

```python
events = events.sort('platform')[:-5]
```

```python
## events by platform
by_platform = events.groupby('platform').uuid.count().reset_index().sort('uuid')
by_platform['platform'] = by_platform.platform.astype('str').str.replace('.0', '')
plt.bar(by_platform.platform, by_platform.uuid)
```

```python
## Catgory of users, example - count of users who appeared 10 times, 20 times, 50 times
users_by_doc = events.groupby('uuid')['uuid'].count()

plt.figure(figsize=(12, 8))
plt.hist(users_by_doc.values, bins=50, log=True)
plt.xlabel('Number of times user appeared', fontsize=12)
plt.ylabel('log(Count of events)', fontsize=12)
plt.show()    
```

```python
##frequency of ads in training data
grouped_by_ad = train.groupby('ad_id').display_id.count()
for i in [2, 10, 50, 100, 1000]:
    print('Ads that appear less than {} times: {}%'.format(i, round((grouped_by_ad < i).mean() * 100, 2)))

plt.figure(figsize=(12, 6))
plt.hist(grouped_by_ad.values, bins=50, log=True)
plt.xlabel('Number of times Ad appeared', fontsize=12)
plt.ylabel('log(Count of displays with ad)', fontsize=12)
plt.show()    
```
